## General Motors analiza actiuni GM (4 luni) de la inceputul anului


```python
import pandas_datareader as pdr
import datetime

gm = pdr.get_data_yahoo('GM',
                       datetime.datetime(2020, 1, 3),
                       datetime.datetime(2020, 4, 8))

# gm.head()
# gm.tail()
gm.describe()
```
